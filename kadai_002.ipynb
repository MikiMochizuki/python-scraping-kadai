```python
import requests
import pprint
from getpass import getpass
```

```python
# APIキーを指定
api_key = getpass('APIキーを入力してください: ')
```

```python
# 検索したい駅名をstation_nameに格納
station_name = input('検索する駅を入力してください: ')
```

```python
# Google Places APIのエンドポイントを定義
url =  'https://maps.googleapis.com/maps/api/geocode/json'
response = requests.get(
    url,
    params={
         'key': api_key,  # APIキーをパラメータに追加
         'address': station_name  #駅名を追加
   }
)
# レスポンスのJSONを取得
json_data = response.json()

# JSONの中身を確認
pprint.pprint(json_data)
```

```python
# Google Places APIのエンドポイントを定義
url2 = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"  # ← ここを追加

response2 = requests.get(
    url2,

# リクエストに渡すパラメータ
   params = {
        'location': '35.6996473,139.7713703',  # 緯度と経度を1つのパラメータで指定
        'radius': 500,                        # 半径（メートル）
        'type': 'restaurant',                      # 場所の種類（例: restaurant, cafe, etc.）
        'key': api_key,                          # APIキー
        'language': 'ja'                     # 言語
    }
)
# レスポンスのJSONを取得
json_data2 = response2.json()

# JSONの中身を確認
pprint.pprint(json_data2)

```

```python
# 取得した結果のレストラン名・評価・住所を取得
for item in json_data2['results']:
    name = item.get('name', 'N/A')         # レストラン名を取得（ない場合は 'N/A' を表示）
    rating = item.get('rating', 'N/A')     # 評価を取得（ない場合は 'N/A' を表示）
    vicinity = item.get('vicinity', 'N/A') # 住所を取得（ない場合は 'N/A' を表示）

    print(f' {name}, {rating}, {vicinity}')
```
